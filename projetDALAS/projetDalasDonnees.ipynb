```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 配置 ChromeDriver 路径（请确保已下载相应版本的 chromedriver）

driver = webdriver.Chrome(service=Service(driver_path))

# 定义要爬取的年份区间
start_year = 2001
end_year = 2020
years = range(start_year, end_year + 1)

# 用于存储每一年前 100 名选手姓名的字典，键为年份，值为姓名列表
ranking_data = {}

for year in years:
    # 构造对应年份十二月的页面 URL
    url = f"https://ranking.ittf.com/#/rankings/list/SEN/M;SINGLES/{year}/12"
    driver.get(url)
    
    # 等待页面加载完成，等待页面中包含排名数据的 <table> 元素出现
    try:
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH, "//div[@class='rank-list']/ul"))
        )
    except Exception as e:
        print(f"{year} 年页面加载失败：", e)
        continue

    # 为确保数据加载完毕，可以再等待几秒钟
    time.sleep(2)
    
    # 根据页面实际结构定位所有排名行，这里假定数据位于 table > tbody > tr 中
    rows = driver.find_elements(By.XPATH, "//div[@class='rank-list']/ul//li")
    print(len(rows))
    players = []
    # 取前 100 行，如果不足 100 行则按实际数量
    for i, row in enumerate(rows):
        if i >= 100:
            break
        # 采用相对 XPath 定位单元格，假定姓名在第二个 <td> 元素中（索引为 1）
        cell = row.find_element(By.XPATH, "./div/div[@class='rank-player__name']/div[1]/a/span[2]")
        name = cell.text.strip()
        players.append(name)

    ranking_data[year] = players
    print(f"{year} 年数据爬取完毕，共获得 {len(players)} 个选手姓名。")

driver.quit()

# 整理数据：创建一个 DataFrame，列为年份，行索引为排名（Rank 1 至 Rank 100）
df = pd.DataFrame(ranking_data)
df.index = [f"Rank {i+1}" for i in range(df.shape[0])]

# 输出结果到控制台，并保存为 Excel 文件（或 CSV 文件，根据需要）
print(df)
df.to_csv("ittf_ranking.csv", index=True)
```
